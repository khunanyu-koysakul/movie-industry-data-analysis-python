## Project Objective
This project analyzes movie industry data to identify the key factors affecting box office revenue and investment efficiency.  
The goal is to understand how budget, audience engagement, and release timing influence financial performance.


### Problem Statement

The movie industry requires high production costs, but spending more money does not always guarantee success.

Studios need to decide how much to invest, what type of movies to produce, and when to release them.<br><br>

This project analyzes movie data to understand which factors such as budget, ratings, and production companies are related to box office revenue.

The goal is to turn raw data into useful information that supports better decision making.


### Assumptions

The dataset reflects common trends in the movie industry.


Budget and revenue values are recorded consistently and can be compared across movies.


Because movie earnings are highly uneven, data transformation helps show clearer relationships.


The results show patterns and relationships, not direct cause-and-effect conclusions.

### Results

The analysis shows a clear positive relationship between movie budget and box office revenue.
Movies with larger budgets usually earn more, but the increase is not perfectly proportional — high spending improves the chance of success but does not ensure profit.<br><br>

The data also shows that only a small number of movies generate a large portion of total revenue, meaning the industry depends heavily on major hits.
Certain production companies repeatedly appear among high-earning movies, suggesting consistent performance rather than luck.<br><br>

Ratings have a weaker relationship with revenue compared to budget, indicating that exposure and marketing reach matter more than audience score alone.

### Conclusion

Overall, box office performance is mainly influenced by investment size and market reach rather than quality indicators alone.
Large-budget movies tend to produce stable results, while smaller movies carry higher risk but sometimes deliver very high returns.

This suggests the movie industry works best as a portfolio strategy , combining reliable big productions with selective smaller projects to balance risk and reward.

### Suggestions / Recommendations

Based on the analysis, the following suggestions can be derived:

1. Focus on Budget Allocation
Since budget shows a strong positive relationship with gross revenue, allocating sufficient budget—especially for marketing and production quality—can increase the likelihood of higher box office returns.

2. Leverage Successful Production Companies
A small number of production companies consistently dominate high-grossing movies. Collaborating with experienced and successful studios may improve commercial performance.

3. Do Not Rely Solely on Ratings
Movie ratings show a weaker relationship with revenue compared to budget. While quality remains important, ratings alone should not be the primary predictor of financial success.

4. Future Analysis Opportunities
Future studies could include additional factors such as genre, release date, or marketing spend to gain deeper insights into box office performance.

Average Gross by Budget Range (Investment Efficiency)

This chart compares average revenue across budget groups to evaluate investment efficiency and shows diminishing returns at very high budget levels.

```python
import pandas as pd
import matplotlib.pyplot as plt

bins = [0, 20000000, 50000000, 100000000, 200000000, 500000000]
labels = ['<20M', '20-50M', '50-100M', '100-200M', '200M+']

df['budget_group'] = pd.cut(df['budget'], bins=bins, labels=labels)

avg_gross = df.groupby('budget_group')['gross'].mean()

plt.figure(figsize=(8,5))
avg_gross.plot(kind='bar')

plt.title('Average Gross by Budget Range')
plt.xlabel('Budget Range')
plt.ylabel('Average Gross Earnings')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Profitability Distribution (ROI Analysis)

The histogram illustrates the distribution of profitability across movies, showing that most films achieve moderate returns while a small number generate extremely high ROI, reflecting a hit-driven market structure.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Project/movies.csv")   
df['ROI'] = df['gross'] / df['budget']


df = df.replace([float('inf'), -float('inf')], None)
df = df.dropna(subset=['ROI'])


import numpy as np

plt.figure(figsize=(8,5))
plt.hist(np.log1p(df['ROI']), bins=50)

plt.title('ROI Distribution (Log Scale)')
plt.xlabel('Log ROI')
plt.ylabel('Number of Movies')
plt.show()
```

Average ROI by Genre

This visualization compares profitability across genres, showing that financial performance varies by category rather than scale alone.

```python
import pandas as pd
import matplotlib.pyplot as plt

# read data
df = pd.read_csv("C:/Project/movies.csv")

# create ROI
df['ROI'] = df['gross'] / df['budget']

# remove invalid values
df = df.replace([float('inf'), -float('inf')], None)
df = df.dropna(subset=['ROI'])

# groupby
genre_roi = df.groupby('genre')['ROI'].mean().sort_values(ascending=False)

# plot
genre_roi.plot(kind='bar', figsize=(10,5))
plt.title('Average ROI by Genre')
plt.xlabel('Genre')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.show()
```

Release Month

This analysis evaluates seasonal performance patterns, showing that movie revenue varies depending on release timing.

```python
import pandas as pd
import matplotlib.pyplot as plt

# แปลงเป็น string
df['released_str'] = df['released'].astype(str)

# หาเดือนจากข้อความ
months = ['January','February','March','April','May','June',
          'July','August','September','October','November','December']

def find_month(text):
    for i, m in enumerate(months, 1):
        if m in text:
            return i
    return None

df['month'] = df['released_str'].apply(find_month)

# คำนวณรายได้เฉลี่ย
monthly = df.groupby('month')['gross'].mean()

# plot
plt.figure(figsize=(8,5))
monthly.plot(marker='o')

plt.title('Average Gross by Release Month')
plt.xlabel('Month')
plt.ylabel('Average Gross Revenue')
plt.xticks(range(1,13))
plt.show()
```
